# Checkpoint 1: Data Cleaning 1

Load the companies file into data frames and name it companies.

```python
companies = pd.read_csv('companies.txt',encoding='ISO-8859-1',sep='\t')
#companies.shape
companies.head()
```

Load the rounds data into a data frames and name it rounds2.

```python
rounds2 = pd.read_csv('rounds2.csv',encoding='ISO-8859-1')
rounds2.shape
```

```python
companies.tail()
```

```python
companies['permalink'] = companies[['permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
companies['name'] = companies.name.str.encode('utf-8').str.decode('ascii', 'ignore')
companies['permalink'] =  companies['permalink'].apply(lambda x: x.lower()).str.split('/').str[2]
companies['name'] = companies['name'].str.lower()
```

```python
companies.tail()
```

```python
rounds2.tail()
```

```python
rounds2['company_permalink'] =  rounds2[['company_permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
rounds2['company_permalink'] =  rounds2['company_permalink'].apply(lambda x: x.lower()).str.split('/').str[2]
```

###### A plot showing the number of investments in the top 3 sectors of the top 3 countries on one chart (for the chosen investment type FT).
